```python
from sqlalchemy import create_engine
import sqlalchemy
import pandas as pd
```

```python
engine = create_engine(conn_str)
```

```python
def rentals_may (engine):
    return pd.read_sql("""select 
	*
from 
	sakila.rental r 
where 
	r.rental_date between '2005-05-01' and '2005-05-31'""", engine)
```

```python
df_rentals_may = rentals_may(engine)
df_rentals_may
```

```python
def rentals_june (engine):
    return pd.read_sql("""select 
	*
from 
	sakila.rental r 
where 
	r.rental_date between '2005-06-01' and '2005-06-30'""", engine)
```
